```python
import matplotlib.pyplot as plt
import pandas as pd

data1 = pd.read_csv("Spotify Project/Listening Time per Day of the Week.csv")
dw = pd.DataFrame(data1)

X = list(dw.iloc[:, 0])
Y = list(dw.iloc[:, 1])


plt.figure(figsize=(10, 7))
plt.bar(X, Y, color='r', ec='k', lw=1)
plt.xlabel("Day of the Week")
plt.ylabel("Yearly Minutes Played")
plt.title("What Time of Day Do I Listen to Music?")

plt.show()
```

```python
data2 = pd.read_csv("Spotify Project/Listening Time per Month.csv")
dw = pd.DataFrame(data2)

X = list(dw.iloc[:, 0])
Y = list(dw.iloc[:, 1])


plt.figure(figsize=(10, 7))
plt.plot(X, Y, color='b')
plt.xlabel("Month-Year")
plt.ylabel("Hours Listened To")
plt.title("How Many Hours Have I Listened Per Month?")

plt.show()
```

```python
data3 = pd.read_csv("Spotify Project/AvgHourlyListening Time.csv")
dw = pd.DataFrame(data3)

X = list(dw.iloc[:, 2])
hrs = list(dw.iloc[:, 0])

plt.figure(figsize=(10, 10))
plt.pie(X)
plt.legend(title = "Hour No. of the Day", labels = hrs)
plt.title("Percentage of Minutes Listened to in Different Times of the Day")

plt.show()
```

```python
data4 = pd.read_csv("Spotify Project/DailyListeningTime.csv")
dw = pd.DataFrame(data4)

X = list(dw.iloc[:, 0])
Y = list(dw.iloc[:, 1])

plt.figure(figsize=(20, 5))
plt.plot(X, Y, c='green')
plt.ylabel("Minutes Played per Day")
plt.title("Listening history from 10/10/24 to 10/10/25")
plt.xticks([])

plt.show()
```
